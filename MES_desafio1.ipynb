```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt

a = 4
b = 2
c = 1

def f(x,y,z):
    return (x/a)**2 + (y/b)**2 + (z/c)**2

V_e = 4/3*np.pi*a*b*c   # Volume do elipsoide
V_p = 2*a * 2*b * 2*c   # Volume do paralelep√≠pedo

N = 10**5 # n√∫mero de pontos a amostrar

# gerar tr√™s arrays de aleat√≥rios de tamanho N
r1 = -a + 2*a*rnd.random(N)
r2 = -b + 2*b*rnd.random(N)
r3 = -c + 2*c*rnd.random(N)

passam = np.zeros(N)
passam[f(r1,r2,r3) <= 1] = 1

V_mc = np.sum(passam)/N * V_p

print('Volume do elips√≥ide: {:.5}'.format(V_e))
print('Volume estimado po MC: {:.5}'.format(V_mc))
print('Erro relativo (%): {:.3}'.format(100 * np.abs(V_e - V_mc)/V_e))

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(r1, r2, r3, color = 'r', s = 10, alpha = 0.01)
ax.scatter(r1[passam == 1], r2[passam == 1], r3[passam == 1], color = 'g', s = 10, alpha = 0.1)
plt.show()
```
