```python
# Pontos a gerar
Np = int(2e4)

# arrays para as coordenadas
x = np.zeros(Np)
y = np.zeros(Np)
z = np.zeros(Np)
```

```python
# Como não fazer -- 1
# Gerar pontos
for i in range(Np):
    x0 = rnd.uniform(-1, 1)
    y0 = rnd.uniform(-1, 1)
    z0 = rnd.uniform(-1, 1)
    length = np.sqrt(x0**2 + y0**2 + z0**2)
    x[i] = x0/length
    y[i] = y0/length
    z[i] = z0/length
```

```python
# Como não fazer -- 2
# Gerar pontos
for i in range(Np):
    theta = rnd.uniform(0, np.pi)
    phi = rnd.uniform(0, 2*np.pi)
    x[i] = np.sin(theta)*np.cos(phi)
    y[i] = np.sin(theta)*np.sin(phi)
    z[i] = np.cos(theta)
```

```python
# Como fazer
for i in range(Np):
    cos_theta = rnd.uniform(-1, 1)
    theta = np.arccos(cos_theta)

    phi = rnd.uniform(0, 2*np.pi)
    x[i] = np.sin(theta)*np.cos(phi)
    y[i] = np.sin(theta)*np.sin(phi)
    z[i] = np.cos(theta)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z, marker='.',s=1,c='k') # esfera unitária
display(fig)
plt.show()
```
