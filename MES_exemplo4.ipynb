```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)/x

i_ana = 1.851937   # integral anal√≠tico

N = 10**4 # n√∫mero de pontos a amostrar

# gerar dois arrays de aleat√≥rios de tamanho N
r1 = np.pi*rnd.random(N)  # o eixo xx varia entre 0 e Pi
r2 = rnd.random(N)        # o eixo yy varia entre 0 e 1

passam = np.zeros(N)
passam[r2 < f(r1)] = 1

i_mc = np.sum(passam)/N * np.pi

print('Integral anal√≠tico: {:.5}'.format(i_ana))
print('Integral de MC: {:.5}'.format(i_mc))
print('Erro relativo (%): {:.3}'.format(100 * np.abs(i_ana - i_mc)/i_ana))

plt.figure(1)
plt.plot(r1, r2, 'ro', markersize=0.5)
plt.plot(r1[passam == 1], r2[passam == 1], 'go', markersize=0.5)
x = np.arange(0.01, np.pi, 0.01)
plt.plot(x, f(x), 'k', linewidth=2)
plt.show()
```
