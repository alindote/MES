```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt


N = 10**6 # número de pontos a amostrar

x_uni = rnd.random(N)       # gerar um array de aleatórios de tamanho N
x_exp = -1*np.log(1-x_uni)  # para cada aleatório, obter o correspondente na distribuição exponencial

plt.figure(1)
plt.subplot(211)
plt.hist(x_uni, bins=40, density=True, alpha=0.5, edgecolor='black')
plt.subplot(212)
plt.hist(x_exp, bins=40, density=True, alpha=0.5, edgecolor='black')
plt.show()
```
