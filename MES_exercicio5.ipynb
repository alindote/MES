```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
N = 10**6 # número de pontos a amostrar

x_uni = rnd.random(N)       # gerar um array de aleatórios de tamanho N
x_px = np.sqrt(x_uni/(1-x_uni))  # para cada aleatório, obter o correspondente na CDF invertida

plt.figure(1)
plt.subplot(211)
plt.hist(x_uni, bins=40, density=True, alpha=0.5, edgecolor='black')
plt.subplot(212)
plt.hist(x_px, bins=100, range=[0,10], density=True, alpha=0.5, edgecolor='black')
plt.show()
```
