```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
# Valores analiticos:
# integral_0^π sin^2(x) dx = π/2≈1.5708
v1 = np.pi/2.0
# integral_(-1)^1 sqrt(1 - x^2)dx = π/2≈1.5708
v2 = np.pi/2.0
# integral_0^π integral_0^π sin(y) cos^2(x) dx dy = π≈3.14159
v3 = np.pi
# integral_0^1 integral_0^1 integral_0^1 x^2 y z^3 dx dy dz = 1/24≈0.0416667
v4 = 1.0/24.
```

```python
def f1(x):
    return (np.sin(x))**2

def f2(x):
    return np.sqrt(1-x**2)

def f3(x,y):
    return np.sin(y)*(np.cos(x))**2

def f4(x,y,z):
    return x**2*y*z**3
```

```python
# Arrays para representar a diferença vs amostras
Ncs = [500, 1000, int(1e4), int(1e5), int(1e6)]
```

```python
# Função 1

x_plot = []
y_plot = []

for n in Ncs:
  # gerar um array de aleatórios de tamanho Ncs
  r1 = np.pi*rnd.random(n)  # a integração é entre 0 e Pi

  # obter o array com os valores da função nos pontos aleatórios
  fvals = f1(r1)

  F_mc = np.pi*fvals.sum()/n

  # preencher os arrays para o plot
  x_plot.append(np.log10(n))
  y_plot.append(np.log10(np.abs(v1 - F_mc)))
  print('Pontos amostrados: ', n)
  print('Integração analítica: {:.5}'.format(v1))
  print('Integração estimada po MC: {:.5}'.format(F_mc))
  print('Erro relativo (%): {:.3}'.format(100 * np.abs(v1 - F_mc)/v1))

# fazer o plot do erro
fig,ax = plt.subplots()
ax.plot(x_plot, y_plot, 'k', linewidth=2)
ax.set_xlabel(r"$\log_{10}(N_c)$")
ax.set_ylabel(r"$\log_{10}(f_i(x) - v_i)$")
plt.show()
```

```python
# Função 3

x_plot = []
y_plot = []

for n in Ncs:
  # gerar um array de aleatórios de tamanho Ncs
  r1 = np.pi*rnd.random(n)  # a integração é entre 0 e Pi
  r2 = np.pi*rnd.random(n)  # a integração é entre 0 e Pi

  # obter o array com os valores da função nos pontos aleatórios
  fvals = f3(r1,r2)

  F_mc = np.pi**2*fvals.sum()/n

  # preencher os arrays para o plot
  x_plot.append(np.log10(n))
  y_plot.append(np.log10(np.abs(v3 - F_mc)))
  print('Pontos amostrados: ', n)
  print('Integração analítica: {:.5}'.format(v3))
  print('Integração estimada po MC: {:.5}'.format(F_mc))
  print('Erro relativo (%): {:.3}'.format(100 * np.abs(v3 - F_mc)/v3))

# fazer o plot do erro
fig,ax = plt.subplots()
ax.plot(x_plot, y_plot, 'k', linewidth=2)
ax.set_xlabel(r"$\log_{10}(N_c)$")
ax.set_ylabel(r"$\log_{10}(f_i(x) - v_i)$")
plt.show()
```
