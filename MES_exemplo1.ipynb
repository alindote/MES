```python
import numpy as np
import matplotlib.pyplot as plt

a = 7**5
m = 2**31 - 1
I = 23764823

random = []

for i in range(10**5):
  I = np.mod(I*a,m)
  random.append(I/m)
  #print('Número pseudo-aleatório em [0,1[ {:.5}'.format(I/m))

plt.figure()
plt.hist(random, bins=20)
plt.show()
```
