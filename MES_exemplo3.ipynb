```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt


N = 10**6 # número de pontos a amostrar

# gerar dois arrays de aleatórios de tamanho N
r1 = rnd.random(N)
r2 = rnd.random(N)

# para cada par de aleatórios, obter o correspondente na distribuição gaussiana
x_gauss = np.sqrt(-2*np.log(r1))*np.cos(2*np.pi*r2)

# opcional: personalizar para a média e desvio padrão pretendidos
#x_gauss = 3 + x_gauss*1.5

plt.figure(1)
plt.subplot(211)
plt.hist(r1, bins=40, density=True, alpha=0.5, edgecolor='black')
plt.subplot(212)
plt.hist(x_gauss, bins=40, density=True, alpha=0.5, edgecolor='black')
plt.show()
```
