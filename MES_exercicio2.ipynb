```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
```

```python
def f1(x):
  return np.exp(-1*x**2)

def f2(x):
  return 1/(x**2+np.cos(x)**2)
```

```python
# Célula auxiliar para representar as funções
x = np.arange(0.01, np.pi, 0.01)

plt.figure(1)
plt.plot(x, f2(x))
plt.show()
```

```python
# Integral de f1

N = 10**4 # número de pontos a amostrar

# gerar dois arrays de aleatórios de tamanho N
r1 = rnd.random(N)  # o eixo xx varia entre 0 e 1
r2 = rnd.random(N)        # o eixo yy varia entre 0 e 1

passam = np.zeros(N)
passam[r2 < f1(r1)] = 1

i_mc = np.sum(passam)/N

print('Integral de MC: {:.5}'.format(i_mc))

plt.figure(1)
plt.plot(r1, r2, 'ro', markersize=0.5)
plt.plot(r1[passam == 1], r2[passam == 1], 'go', markersize=0.5)
x = np.arange(0.01, 1, 0.01)
plt.plot(x, f1(x))
plt.show()
```

```python
# Integral de f2

N = 10**4 # número de pontos a amostrar

# gerar dois arrays de aleatórios de tamanho N
r1 = np.pi*rnd.random(N)  # o eixo xx varia entre 0 e 1
r2 = rnd.random(N)        # o eixo yy varia entre 0 e 1

passam = np.zeros(N)
passam[r2 < f2(r1)] = 1

i_mc = np.sum(passam)/N * np.pi

print('Integral de MC: {:.5}'.format(i_mc))

plt.figure(1)
plt.plot(r1, r2, 'ro', markersize=0.5)
plt.plot(r1[passam == 1], r2[passam == 1], 'go', markersize=0.5)
x = np.arange(0.01, np.pi, 0.01)
plt.plot(x, f2(x))
plt.show()
```
